```python
import pandas as pd
import matplotlib.pyplot as plt
with open("/content/Queryr.csv","r") as f: #openning file Queryr as a file f
  df=pd.read_csv(f,names=['DATE','TAG','POSTS'],header=0)
  print("Before changing the time \n")
  print(df)
  df.DATE=pd.to_datetime(df.DATE) #Pandas to_datetime() method helps to convert string Date time into Python Date time object.
  print(df)
  reshaped_df=df.pivot(index='DATE',columns='TAG',values='POSTS')
  print(reshaped_df)
```

```python
reshaped_df.fillna(0,inplace=True) #The fillna() method replaces the NULL values with a specified value, here 0
print(reshaped_df)
```

```python
plt.plot(reshaped_df.index,reshaped_df.java)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
```

```python
plt.figure(figsize=(7,5))
plt.plot(reshaped_df.index,reshaped_df.python)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
```

```python
plt.figure(figsize=(7,5)) # changing figure size
plt.plot(reshaped_df.index,reshaped_df.r)
plt.plot(reshaped_df.index,reshaped_df.python)
plt.plot(reshaped_df.index,reshaped_df.c)
plt.xlabel('Date', fontsize=14) # x axis
plt.ylabel('Number of Posts', fontsize=14) #y axis
plt.ylim(0, 35500) 
```

```python
plt.figure(figsize=(14,10)) # changing figure size
for i in reshaped_df.columns: #adding for loop to print all data together
  plt.plot(reshaped_df.index,reshaped_df[i],linewidth=3,label=reshaped_df[i].name)
  plt.xlabel('Date', fontsize=14) # x axis
  plt.ylabel('Number of Posts', fontsize=14) #y axis
  plt.legend(fontsize=14) #adding legends
```
